```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
```

```python
train_ch1 = pd.read_csv('./vibration_data/train_after_ch1.csv', header= None)
train_ch2 = pd.read_csv('./vibration_data/train_after_ch2.csv', header= None)
train_ch3 = pd.read_csv('./vibration_data/train_after_ch3.csv', header= None)
```

```python
print(train_ch1.shape)
train_ch1.head()
```

```python
train_ch1[0]
```

```python
fig = plt.figure(figsize=(17,15))
num_row = 4
num_clm = 3
for i in range(num_row):
    ax = fig.add_subplot(num_row, num_clm, i*num_clm+1)
    ax.plot(train_ch1[i])
    ax.set_title(str(i) + " : ch1")
    ax.set_ylim([-5, 5])

    ax = fig.add_subplot(num_row, num_clm, i*num_clm+2)
    ax.plot(train_ch2[i])
    ax.set_title(str(i) + " : ch2")
    ax.set_ylim([-5, 5])

    ax = fig.add_subplot(num_row, num_clm, i*num_clm+3)
    ax.plot(train_ch3[i])
    ax.set_title(str(i) + " : ch3")
    ax.set_ylim([-5, 5])
```

```python

f, t, Sxx = signal.spectrogram(train_ch1[0])

plt.pcolormesh(t, f, 10*np.log(Sxx)) #intensity
plt.ylabel('Frequency [Hz]')
cbar = plt.colorbar() #
cbar.ax.set_ylabel("Intensity [dB]") #
plt.show()
```
